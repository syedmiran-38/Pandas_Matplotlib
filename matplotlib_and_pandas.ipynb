# **REAL WORLD EXAMPLES USING PANDAS AND MATPLOTLIB**

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

# **Line Graphs**

```python
gas = pd.read_csv('/content/gas_prices.csv')

gas.head()
```

```python
plt.figure(figsize=(10,7))

plt.title('Gas prices over time in USD', fontdict = {'fontweight' : 'bold', 'fontsize' : 20})

plt.plot(gas.Year, gas.USA, label = 'USA')
plt.plot(gas['Year'], gas['Canada'], label = 'Canada')
plt.plot(gas.Year, gas.Australia, label = 'Australia')

plt.xticks(gas.Year[::3].tolist() + [2011])

plt.xlabel('Year')
plt.ylabel('USD')

plt.legend()

plt.show()
```

# **Fifa DataSet**

```python
fifa = pd.read_csv('/content/fifa_data.csv')
fifa
```

# **Histograms**

```python
plt.figure(figsize=(14,9))

bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins = bins, color = 'r')

plt.xticks(bins)

plt.ylabel('Number of players')

plt.xlabel('Skill level')

plt.title('Distribution of players skills in Fifa 18', fontdict = {'fontweight' : 'bold', 'fontsize' : 20})



plt.show()
```

# **PIE CHART**

```python
fifa.head(5)
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']

colors = ['yellow', 'green']

plt.pie([left,right], labels = labels, colors = colors, autopct='%.2f %%')

plt.title('Foot preference of Fifa players')
plt.show()
```

# **Advance Pie Chart**

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

plt.figure(figsize=(20,14))

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]

light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]

medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]

medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]

heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]

labels = ['Under 125', '125-150', '150-175', '175-200', 'over 200']

explode = (.4, .2, 0, 0, .4)

plt.title('Weight distribution of Fifa players in Lbs')

plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance = 0.8, explode = explode)

plt.show()
```

# **Whiskers**

```python
plt.style.use('default')

plt.figure(figsize=(8,11))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']

madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']

revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC barcelona', 'Real Madrid', 'New England Revolution']

plt.title('Professional soccer team comparision')
plt.ylabel('Fifa overall rating')

boxes = plt.boxplot([barcelona, madrid, revs], labels = labels, patch_artist = True, medianprops = {'linewidth' : 2})

for box in boxes['boxes']:
    box.set(color = 'green', linewidth = 2)

    box.set(facecolor = 'yellow')

plt.show();
```
